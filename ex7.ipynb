```python
#How do i sort a pandas DataFrame or Series?
```

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies['title'].sort_values(ascending=False)
```

```python
movies.sort_values('duration',ascending=False)
```

```python
movies.sort('title')
```
